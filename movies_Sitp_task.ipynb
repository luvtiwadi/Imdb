```python
print("Ans 22")
from collections import Counter
data_count=Counter()
for i,j in zip(df['Actors'],df['Director']):
    l1=[]
    actors_list=i.split(',')
    for k in actors_list:
        l1.append(k+","+j)
    data_count.update(l1)
    l1.clear()
most_frequrnt_pair_actor_director=data_count.most_common(3)
print("Most frequrnt pair of actor and director\n")
for i in most_frequrnt_pair_actor_director:
    l=i[0].split(',')
    print(f"{l[0]} -- {l[1]} ==> {i[1]}")
#     print(l[0],"  ",l[1],"==>",i[1])
```

```python
print("Ans 24")
genre_data=df.groupby(['Year','Genre'])['Genre'].count()
genre_data=pd.DataFrame(genre_data)
genre_data.rename(columns={'Genre':'Count'},inplace=True)
genre_data.reset_index(inplace=True)
genre_data
```

```python
print("Ans 24")
pivot_table = genre_data.pivot_table(index='Year', columns='Genre', values='Count', fill_value=0)
pivot_table.plot(kind='barh', stacked=True, figsize=(20, 8))
plt.xlabel('Genre Count')
plt.ylabel('Year')
plt.title('Genre Trends Over the Years')
plt.legend(title='Genres')
plt.show()
```
